## 5. Variáveis

Uma variável é um nome que se refere a um valor.

Variáveis são usadas para guardarmos valores que serão usados mais tarde no programa.

## 6. Atribuição

Um comando é uma unidade de código que pode ser executado. 

Um comando de atribuição tem a seguinte forma:

variável = expressão

Significado:
1. O valor da expressão é calculado;
2. A variável é criada, com as características necessárias para referenciar o valor da expressão;
3. A variável passa a fazer referência/guardar o valor

Exemplo:
`a = 5 + 2`
1. Calcula quanto é 5 + 2
2. Cria um espaço na memória para salvar a variável `a`
3. Guarda em `a` o valor de 5 + 2

## 10. Operadores Relacionais 

#### Operadores relacionais são usados para comparar duas variáveis
- `a == b`: a igual a b 
- `a > b`: a maior que b
- `a < b`: a menor que b
- `a >= b`: a maior ou igual a b
- `a <= b`: a menor ou igual a b
- `a != b`: a diferente de b 

Note que `a == b` é diferente de `a = b`, pois `a = b` é um comando de atribuição e `a == b` é um comando de comparação

CUIDADO: Não é possível fazer uma comparação do tipo `>`, `<`, `>=`, `<=` entre, por exemplo, uma string e um inteiro. 

Veja alguns exemplos em que obtemos um erro quando tentamos comparar tipos de variáveis diferentes:

````
[1] < {1:4} #Comparação entre lista e dicionário
````

````
'string' >= 3 #Comparação entre string e inteiro
````

````
list([1]) <= tuple([1])
````

````
float(3.1415) < int(5) #Comparações entre float e int dão certo!
````

````
"Amanda Locatelli" < "amanda locatelli", "Ana Tamais" >= "ana tamais" #Comparações entre string dão certo!
````

## 11. Operadores Lógicos
#### São três os operadores lógicos: `and`, `or` e `not`.

O significado desses operadores é similar ao significado deles em português.

Exemplos:
- `x > 0 and x < 10` é verdadeiro somente se x é maior do que 0 e, ao mesmo tempo, x é menor do que 10. 
- `n % 2 == 0 or n % 3 == 0` é verdadeiro se pelo menos uma das condições é verdadeira, isto é, se o número é divisível por 2 ou divisível por 3.
- O operador `not` nega uma expressão booleana, de forma que `not x > y` é verdadeiro se `x > y` é falso, isto é, se x é menor ou igual a y.

````
1<2 or 2<3 #O comando "or" é True se as duas comparações forem True, ou seja, o "or" não é exclusivo
````

````
1 not in [1, 2, 3]
````

````
(1+2+3)/6 <= 1.5 < 3 and 1 == abs(-1) #A função abs() devolve o módulo do número que for colocado dentro dos parênteses
````

## 12. Atribuição múltipla, de tuplas, e resumida

- **Atribuição Múltipla**: é possível atribuir um valor a várias variáveis num só comando de atribuição. 

    - A expressão à direita é calculada em primeiro lugar e, em seguida, é atribuída a cada elemento

````
a = b = 1 # a e b são iguais a 1
a = b = c = a + b # a, b e c são iguais a a + b, que é 2
print("O valor de a é: {}".format(a))
print("O valor de b é: {}".format(b))
print("O valor de c é: {}".format(c))
````

- **Atribuição de Tuplas**:  é possíveis atribuir diversos valores a várias variáveis num só comando de atribuição

````
a, b, c = "string a", True, 3.2
print("O valor de a é: {}".format(a))
print("O valor de b é: {}".format(b))
print("O valor de c é: {}".format(c))
````

````
print("Antes: (a, b) = ({},{})".format(a, b))
a, b = b, a #Para trocar o valor de duas variáveis a e b entre si
print("Depois: (a, b) = ({},{})".format(a, b))
````

#### 14.9 Funções `lower()`, `upper()`, `capitalize()`, `swapcase()`, e `title()`
- `lower()`: deixa todas letras da string minúsculas
- `upper()`: deixa todas as letras da string maiúsculas
- `capitalize()`: deixa apenas a primeira letra da string maiúscula
- `swapcase()`: troca maiúsculo por minúsculo e vice-versa
- `title()`: deixa a primeira letra de cada palavra da string maiúscula e o resto minúsculo


````
a = "aaAA bbBB"
print("a = {}".format(a))
print("lower(): {}".format(a.lower())) #lower() deixa todas as letras minúsculas
print("upper(): {}".format(a.upper())) #upper() deixa todas as letras maiúsculas 
print("capitalize(): {}".format(a.capitalize())) #capitalize() deixa apenas a primeira letra maiúscula
print("swapcase(): {}".format(a.swapcase())) #swapcase() troca maiúsculo por minúsculo e vice-versa
print("title(): {}".format(a.title())) #title() deixa a primeira letra de cada palavra da string maiúscula
````

#### 14.10 Funções `islower()`, `isupper()`, `istitle()`, `isalnum()`, `isalpha()`, `isdigit()`, `isspace()`
Essas funções tentam validar uma hipótese e retornam um valor booleano (True ou False).
- `islower()`: retorna True se todas as letras forem minúsculas. Caso contrário, retorna False
- `isupper()`: retorna True se todas as letras forem maiúsculas. Caso contrário, retorna False
- `istitle()`: retorna True se a primeira letra de todas as palavras da string forem maiúsculas (e o resto é minúsculo). Caso contrário, retorna False
- `isalnum()`: retorna True se a string tiver apenas caracteres alfanuméricos. Caso contrário, retorna False
- `isalpha()`: retorna True se a string tiver apenas letras. Caso contrário, retorna False
- `isdigit()`: retorna True se a string tiver apenas números. Caso contrário, retorna False
- `isspace()`: retorna True se a string tiver apenas espaços. Caso contrário, retorna False.

````
print("Todas as letras de '{}' são maiúsculas? {}".format('UPPER', 'UPPER'.isupper()))
print("Todas as letras de '{}' são maiúsculas? {}".format('UpPER', 'UpPER'.isupper()))
print("Todas as letras de '{}' são minúsculas? {}".format('lower', 'lower'.islower()))
print("Todas as letras de '{}' são minúsculas? {}".format('loWer', 'loWer'.islower()))
print("A string '{}' é um título? {}".format('Um Título', 'Um Título'.istitle()))
print("A string '{}' é alfanumérica? {}".format("aa44", "aa44".isalnum()))
print("A string '{}' é alfanumérica? {}".format("aa44$", "aa44$".isalnum()))
print("A string '{}' contém somente letras? {}".format('12a', '12a'.isalpha()))
print("A string '{}' contém somente números? {}".format('123', '123'.isdigit()))
print("A string '{}' contém somente espaços? {}".format('   ', '   '.isspace()))
````

## 19. Comandos `while`, `break` e `continue`

`while` é um laço de repetição, assim como o `for`, que será visto mais adiante.  
Laços são usados para repetições dentro do seu programa.  
O passo a passo é bem simples:  
1. Inicializar contador
2. Determinar o número de repetições que o programa irá realizar
3. Determinar o comando que será repetido.
4. Contabilizar a passagem pelo loop     

![Voltas](http://drive.google.com/uc?export=view&id=1oYm0yKUP24BrcROf639y85RZlHiwK_EE)  

`while` = 'Enquanto' em português
  
Exemplo em 'português' no passo a passo:
  
1. contador = 0
2. `Enquanto` o contador < `5` (número de "voltas" que queremos que nosso programa realize
3. Escrever "Eu amo Python!" (comando)
4. contador + 1
  



  



````
#Exemplo com Python
contador =0
while contador < 5:
  print('Eu amo Python!')
  contador += 1
````

O comando `break` é usado para sair do while enquanto ele ainda está sendo executado. O `break` vai ignorar o que ocorrer debaixo dele.
  
Exemplo em 'português':  

```
contador = 1  
Enquanto o contador for maior que 0:  
  somar 1 no contador
  Se n == 3:  
    break
```
O que aconteceria sem o `break`??


````
contador = 1  
while contador > 0:
  contador += 1
  if contador == 3:
    break  
print('Esse laço foi interrompido no contador = 3. \n\nCaso contrário, ele teria continuado para sempre, \npois o contador será sempre maior que 0.')
````

O `continue` é praticamente o oposto do break.  
Enquanto o `break` pára o programa durante a execução, o `continue` vai continuar o código, mesmo após ele ter sido interrompido de alguma forma.


````
contador = 0
while contador < 5:
  contador = contador + 1   #também pode ser escrito como 'contador += 1'
  if contador == 3:
    continue
  print(contador, end= ' ')
````

````
contador = 0
while contador < 10:
  print(contador)
  contador += 1
  if contador > 6:
    continue                  # O que ocorreria com o break?
  else:
    print('Testando o comando CONTINUE')
   
````

Para executar essa sequência de comandos ou esse programa,

Assim, a estrutura fica da seguinte forma:


````
def nome_funcao():
  #insira aqui alguns comandos
  a = 1
  b = 2
  c = a + b
  print("c = {}".format(c))
nome_funcao() #chama e executa a função "nome_funcao()"
````

Estamos na verdade definindo (`def`) uma função sem parâmetros. Por isso o abre e fecha parêntesis vazio. O programa é simplesmente um comando que chama esta função.
Mas afinal, **o que são parâmetros?**

Aproveitaremos para explicar sobre variáveis locais e globais também. Depois voltamos à ideia de função.

### **Variáveis Locais**

Uma variável criada dentro de uma função, só pode ser usada dentro desta função onde foi criada, ou onde recebeu o primeiro valor. Fora desta função a variável é desconhecida. 

Podemos criar então variáveis com o mesmo nome em funções diferentes e serão variáveis diferentes.

Dizemos que **uma variável criada dentro de uma função é uma variável local a esta função.**

Os argumentos ou parâmetros formais de uma função agem também como variáveis locais a esta função.

### **Variáveis Globais**

**Variáveis Globais são aquelas declaradas no início de um algoritmo, fora da função.** São visíveis, ou seja, podem ser utilizadas no algoritmo principal e por todos os outros subalgoritmos. 

Note, portanto, que **uma variável declarada num nível superior a uma função é conhecida dentro desta função.**

Uma pergunta que pode surgir é: Por que não usamos somente variáveis globais, em vez de parâmetros ou variáveis locais a funções?

A resposta é simples. À medida que os programas se tornam grandes e complexos, **aumenta a chance de haver erros**. Ter acesso a variáveis que não precisam ser usadas dentro de uma função é uma possível fonte de erros. Alterar uma variável indevidamente num programa grande é um erro muitas vezes difícil de ser localizado.

Por esse motivo, não ensinaremos nesse curso sobre como instanciar uma variável global dentro da função usando `global`.

### **Parâmetros**
**Parâmetro é uma variável declarada no cabeçalho da função e tem uso exclusivo dentro do bloco de instrução da mesma.**

A definição dos parâmetros que uma função deve receber obriga o envio dos valores todas as vezes em que a mesma for invocada, do contrário, a função não é invocada.

É possível que seja definido um **parâmetro default**. Nesse caso, não é necessário que esse parâmetro seja passado quando a função é chamada. Porém, se você desejar chamar essa função mudando esse parâmetro com valor default, é possível.

## 24. Funções Recursivas

Recursão é um método de resolução de problemas que envolve quebrar um problema em subproblemas menores e menores até chegar a um problema pequeno o suficiente para que ele possa ser resolvido trivialmente.

Normalmente recursão envolve uma função que chama a si mesma.

Embora possa não parecer muito, a recursão nos permite escrever soluções elegantes para problemas que, de outra forma, podem ser muito difíceis de programar.

Vejamos um exemplo: Desejamos calcular, no primeiro exemplo, o fatorial de n, e, no segundo exemplo, a soma de uma lista de números

````
def fatorial(n):
  fat = 1
  if n < 0:
    return None
  for i in range(1, n+1):
    fat *= i
  return fat
print(fatorial(4))

def fatorial(n):
  if n <= 0:
    return 1
  return n * fatorial(n - 1)
print(fatorial(4))
````

````
#Aqui temos um exemplo SEM USAR RECURSÃO
def listsum(lista):
  soma = 0
  for i in lista:
    soma = soma + i
  return soma
print(listsum([1, 2, 3]))

#USANDO RECURSÃO
def listsum(lista):
  if len(lista) == 1:
    return lista[0]
  else:
    return lista[0] + listsum(lista[1:])
print(listsum([1, 2, 3]))
````

## 25. Módulos

Ao sair e entrar de novo no interpretador Python (Shell Mode), as definições anteriores (funções e variáveis) são perdidas. Portanto, se quiser escrever um programa maior, será mais eficiente usar um editor de texto para preparar as entradas para o interpretador, e executá-lo usando o arquivo como entrada. Isso é conhecido como criar um script. 

Se o programa se torna ainda maior, é uma boa prática dividi-lo em arquivos menores, para facilitar a manutenção. Também é preferível usar um arquivo separado para uma função que você escreveria em vários programas diferentes, para não copiar a definição de função em cada um deles.

Para permitir isso, **o Python tem uma maneira de colocar as definições em um arquivo e então usá-las em um script ou em uma execução interativa do interpretador**. Tal arquivo é chamado de **módulo**; definições de um módulo podem ser importadas para outros módulos, ou para o módulo principal.

Um módulo pode conter tanto instruções executáveis quanto definições de funções e classes. Essas instruções servem para inicializar o módulo. Eles são executados somente na primeira vez que o módulo é encontrado em uma instrução de importação.

A seguir, veremos como importar um módulo e alguns módulos (também chamados de bibliotecas) de Python bastante comuns:

````
import pandas as pd #importa pandas como pd
import numpy as np #importa numpy como np
from math import * #importa tudo que está dentro de math
from math import floor, ceil #importa floor e ceil, que estão dentro de math
import random #importa random
import string #importa string
import time #importa time
import statistics #importa statistics
````

O **módulo math** possui várias funções pré-definidas cujo objetivo é facilitar a programação de cálculos matemáticos.

O comando `dir(math)` devolve uma lista com todas as funções deste módulo:

````
import math
print(dir(math))
print(math.floor(2.3)) #Precisa escrever math.<nome da função> para chamar a função
````

````
from math import floor
print(floor(2.3)) #Não precisa escrever math.<nome da função>, apenas <nome da função>, porque floor foi importado de math
````

O **módulo random** contém funções que geram números aleatórios. 

Na verdade o correto é dizer que geram números pseudoaleatórios, pois são baseados em um algoritmo determinístico — reproduzível e previsível. 

A partir de um valor chamado de semente (seed), os valores gerados são os mesmos. A cada vez que você executar o seu programa, o valor da semente é provavelmente diferente, significando que apesar dos números aleatórios estarem sendo criados usando algoritmos determinísticos, você vai obter um comportamento aleatório cada vez que você executar. 

O valor da semente pode também ser alterado pelo programa.

````
import random
random.seed(43)

print(random.random()) #A função random() retorna um número em ponto flutuante no intervalo [0.0, 1.0)
print(random.randrange(7, 10)) #A função randrange gera um inteiro no intervalo definido pelo menor e maior argumento
print(random.uniform(1, 5)) #A função uniform(a, b) retorna um número em ponto flutuante entre a e b
````

O **módulo time** contém funções que consultam as várias temporizações

````
import time
print(time.time()) #devolve o valor corrente do relógio em segundos referentes a uma base de referência (e.g.01/01/1970).
time.sleep(2) #Deixa o programa suspenso por 2 segundos
print(time.clock()) #Devolve o tempo de CPU consumido até o momento do programa
print(time.process_time()) #Devolve o tempo de CPU + tempo do sistema consumido até o momento pelo sistema
````

## 26. Boas práticas gerais de programação (em Python)
1. Nomeie variáveis de maneira intuitiva. Em um projeto, não chame suas variáveis e funções de `a`, por exemplo. Se ela for uma lista de transaction_ids, chame de list_transactionids, por exemplo.
2. No mesmo sentido do tópico 1, crie métodos e funções com objetivos bem definidos e nomes bastante condizentes com esse objetivo.
2. Evite usar variáveis globais em funções
3. Comente o código:
  - Para comentar 1 linha de código, use `#`
  - Para comentar uma função ou uma classe, use 
  `"""comentario"""` ou `''' comentario '''`
4. Use try/except ou até mesmo if/else para evitar erros no código que podem acabar tornando-o inutilizável, principalmente se estiver em produção
6. Caso esteja fazendo um projeto grande, separe-o em arquivos .py, em que cada arquivo tem um objetivo específico. E todos esses arquivos (com classes e/ou funções), são chamados/importados em um só arquivo (main) principal
7. Faça um código limpo e sem "gambiarras". Se a implementação está complicada demais e não organizada, muito provavelmente há erros no meio.
8. Faça um arquivo de testes (dica: aprenda unittest e programação orientada à testes)
9. Quando estiver fazendo um código em um notebook, não faça células muito grandes nem pequenas demais (~1, 2 linhas). Procure quebrar o código em pequenos blocos.
10. Não faça funções gigantescas. Se você estiver fazendo uma função gigantesca, provavelmente você não usando essa função para um objetivo específico e pontual. Tente quebrá-la em outras funções.
12. Se você estiver trabalhando com numpy.array ou pandas.DataFrame (ou outras estruturas de dados semelhantes), dê preferência a operações vetorizadas ao invés do for/while loop
13. Não sobrescreva builtins
14. Evite fazer muitos níveis de indentação quando possível (recomendação: máximo 3 níveis. 4 se estiver dentro de uma classe)
15. Faça bons prints/logs
15. Caso esteja colocando seu código no github, siga as recomendações da equipe de engenharia para a documentação de repositórios: https://cloudwalk.atlassian.net/wiki/spaces/QA/pages/57606327/Patterns

#### Zen of Python
O Zen of Python é uma coleção de 19 "guiding principles" para escrever bons códigos em Python.

1. Beautiful is better than ugly.
2. Explicit is better than implicit.
3. Simple is better than complex.
4. Complex is better than complicated.
5. Flat is better than nested.
6. Sparse is better than dense.
7. Readability counts.
8. Special cases aren't special enough to break the rules.
9. Although practicality beats purity.
10. Errors should never pass silently.
11. Unless explicitly silenced.
12. In the face of ambiguity, refuse the temptation to guess.
13. There should be one—and preferably only one—obvious way to do it.
14. Although that way may not be obvious at first unless you're Dutch.
15. Now is better than never.
16. Although never is often better than right now.
17. If the implementation is hard to explain, it's a bad idea.
18. If the implementation is easy to explain, it may be a good idea.
19. Namespaces are one honking great idea—let's do more of those!
